```python
from pycocotools.coco import COCO
import requests
import os
import json
import skimage.io as io
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import skimage
import cv2
from skimage.transform import resize
```

```python
output='/content/drive/MyDrive/Deep_learning/HW5/images_test/'
os.chdir(output)
input_json="/content/drive/MyDrive/Deep_learning/HW5/annotations/instances_val2017.json"
class_list = ["car","boat"]
coco = COCO(input_json)
catIds = coco.getCatIds(catNms = class_list)
imgIds = coco.getImgIds( catIds = catIds)
for imgId in imgIds:
    img=coco.loadImgs(imgId)[0]
    with open(img['file_name'], 'wb') as coco_write:			
		     coco_write.write(requests.get(img['coco_url']).content)
    img_file_path = os.path.join(output, img['file_name'])
    im = Image.open(img_file_path)
    if im.mode != 'RGB':
		   im = im.convert (mode="RGB")
     im.save(img_file_path)
```
